```python
results = {}

mot_tracker = Sort() #object tracker

# load models
coco_model = YOLO('yolov8x.pt')
#Create a YOLO object with trained model
!cp /content/gdrive/MyDrive/CIS530_Project/license_plate_detector.pt /content
# Load a model
#license_plate_detector = YOLO('license_plate_detector.pt')  # load an official model
license_plate_detector = YOLO('yolov8n.pt')  # load an official model
license_plate_detector = YOLO('./best.pt')# load a custom model

# Path to the source video file
!cp /content/gdrive/MyDrive/CIS530_Project/sample.mp4 /content
video_path = './sample.mp4'
# load video
cap = cv2.VideoCapture(video_path)

# Check if the video capture is successful
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Directory to save captured images
output_directory = 'captured_images'
os.makedirs(output_directory, exist_ok=True)

#index from https://raw.githubusercontent.com/pjreddie/darknet/master/data/coco.names
vehicles = [2, 3, 5, 7] #represent diff type of ve

# read frames
frame_nmr = -1
ret = True
while ret:
    frame_nmr += 1
    ret, frame = cap.read()

    if ret:
        if frame_nmr > 10:
            break

        results[frame_nmr] = {}
        # detect vehicles
        detections = coco_model(frame)[0]
        detections_ = []
        for detection in detections.boxes.data.tolist():
            x1, y1, x2, y2, detect_score, class_id = detection
            if int(class_id) in vehicles:
                detections_.append([x1, y1, x2, y2, detect_score])

        # track vehicles
        track_ids = mot_tracker.update(np.asarray(detections_))

        # detect license plates
        license_plates = license_plate_detector(frame)[0]
        for license_plate in license_plates.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = license_plate

            # assign license plate to car
            xcar1, ycar1, xcar2, ycar2, car_id = get_car(license_plate, track_ids)

            if car_id != -1:

                # crop license plate
                license_plate_crop = frame[int(y1):int(y2), int(x1): int(x2), :]

                # process license plate
                license_plate_crop_gray = cv2.cvtColor(license_plate_crop, cv2.COLOR_BGR2GRAY)
                _, license_plate_crop_thresh = cv2.threshold(license_plate_crop_gray, 64, 255, cv2.THRESH_BINARY_INV)

                # read license plate number
                license_plate_text, license_plate_text_score = read_license_plate(license_plate_crop_thresh)

                """ cv2.imshow('original_crop', license_plate_crop)
                cv2.imshow('threshold', license_plate_crop_thresh)
                cv2.waitKey(0) """

                if license_plate_text is not None:
                    results[frame_nmr][car_id] = {'car': {'bbox': [xcar1, ycar1, xcar2, ycar2], 'detect_score': detect_score}, # bounding box
                                                  'license_plate': {'bbox': [x1, y1, x2, y2],
                                                                    'text': license_plate_text,
                                                                    'bbox_score': score,
                                                                    'text_score': license_plate_text_score}}
cap.release()
cv2.destroyAllWindows()
# write results
write_csv(results, './test_v8x_best.csv')

```
